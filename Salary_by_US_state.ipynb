In real scenario, we could refer to the database, to correct by filling up.

But if there is none, we need to do data manipulattion. We sould see earlier in . describe fnx
Mean = 1,176,343
Median (50th percentile) = 1,381,701
This indicates that the distribution is skewed by low
Because of these low outliers/zeros, the mean is pulled downward and becomes less representative of a "typical" state's profit.
The median, on the other hand, is more robust to extreme low values. It better reflects the central tendency.

We do the data manipulation

```python

median_profit = data[data['avg_profit'] > 0]['avg_profit'].median()

print(f"Median profit (from non-zero states): {median_profit:.2f}")

data['avg_profit_filled'] = data['avg_profit'].replace(0, median_profit)


data['avg_profit_filled'] = data['avg_profit_filled'].fillna(median_profit)

print("\nBefore filling:")
print(data['avg_profit'].describe().round(2))

print("\nAfter filling with median:")
print(data['avg_profit_filled'].describe().round(2))

print(f"\nNumber of zeros before filling: {(data['avg_profit'] == 0).sum()}")
print(f"Number of zeros after filling: {(data['avg_profit_filled'] == 0).sum()}")
```

But still the state New Mexico has missing value, lets try other approach using lambda as I believe this is betteer approach in manipulating per row

```python
new_mexico
```

```python
median_profit = data[data['avg_profit_filled'] > 0]['avg_profit_filled'].median()
print(f"Median profit to be used for filling: {median_profit:.2f}")

data['avg_profit_filled'] = data['avg_profit_filled'].copy()

data['avg_profit_filled'] = data['avg_profit_filled'].where(data['avg_profit_filled'] > 1, median_profit)
```

```python
median_profit = data[data['avg_profit_filled'] > 0]['avg_profit_filled'].median()

print(f"Median profit used for filling: {median_profit:.2f}")

data['avg_profit_filled'] = data.apply(
    lambda row: median_profit if row['avg_profit_filled'] <= 1 else row['avg_profit_filled'],
    axis=1
)
```

```python
data.tail(10)
```

It appears lambda works, lets remove the old avg_profit as if the row under avg_profi has value will return the same

```python
data = data.drop(columns=['avg_profit'])
```

I decided to usse median as safe option in data manipulation, as you could remmember we have outliers and if we use mean outliers has a big impact in terms of the reliability of used data in manipulation

```python
# Step 1: Calculate the median of total_profit from states that have positive values
median_total_profit = data[data['total_profit'] > 0]['total_profit'].median()

print(f"Median total_profit used for filling: {median_total_profit:.2f}")

# Step 2: Fill using lambda (as you requested)
data['total_profit_filled'] = data.apply(
    lambda row: median_total_profit if row['total_profit'] <= 1 else row['total_profit'],
    axis=1
)

# Step 3: Check the result
print("\nNew Mexico check after filling:")
print(data[data['state_usa'] == 'New Mexico'][['state_usa', 'total_profit', 'total_profit_filled']])

print(f"\nNumber of near-zero total_profit before: {(data['total_profit'] <= 1).sum()}")
print(f"Number of near-zero total_profit after: {(data['total_profit_filled'] <= 1).sum()}")
```

Lets test our observation by doing inferential analysis

we use spearman rho as we are handling ranking which is ordinal

```python
from scipy.stats import spearmanr

# Spearman correlation between population ranking and profit ranking
corr, p_value = spearmanr(data['population_ranking'], data['profit_ranking'])

print(f"Spearman correlation: {corr:.4f}")
print(f"p-value: {p_value:.4f}")
```

**Result**


We found no meaningful correlation between population ranking and profit ranking (Spearman correlation = 0.0035, p-value = 0.9806).
Interpretation

The near-zero correlation coefficient indicates that there is no systematic relationship between a state's population ranking and its profit ranking.
Our initial observation that states with higher population ranking tend to have lower profit ranking appears to be an artifact of looking only at the bottom 5 states. When we analyze the full dataset, this pattern does not hold.
This result suggests that population size alone does not strongly influence profit ranking. Other factors (such as industry composition, economic structure, or business environment) may play a more significant role.
